```python


def scale_image_nearest(image, factor):
    #height, width = image.shape[:2]
    height, width = image.shape
    new_height = int(height * factor)
    new_width = int(width * factor)

    # Erzeuge ein leeres Bild mit der skalierten Größe
    #scaled_image = np.zeros((new_height, new_width, image.shape[2]), dtype=image.dtype)
    scaled_image = np.zeros((new_height, new_width), dtype=image.dtype)

    for i in range(new_height):
        for j in range(new_width):
            # Finde den nächstgelegenen Nachbarn im Originalbild
            orig_i = int(i / factor)
            orig_j = int(j / factor)

            # Weise den Wert des nächstgelegenen Nachbarn dem entsprechenden Pixel im skalierten Bild zu
            scaled_image[i, j] = image[orig_i, orig_j]

    return scaled_image

# Lade das Bild
image = imread('./tv.png')

# Skaliere das Bild um den Faktor 1.5
#scaled_image = scale_image_nearest(image, 1.5)
#scaled_image = scale_image_nearest(np.copy(image), 5.0)
scaled_image = scale_image_nearest(image, 5.0)

# Zeige das Originalbild und das skalierte Bild an
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Originalbild')

plt.subplot(1, 2, 2)
plt.imshow(scaled_image)
plt.title('Skaliertes Bild')

```
