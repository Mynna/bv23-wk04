```python
import numpy as np
import skimage.io
import skimage.filters
import skimage.color
import matplotlib.pyplot as plt



# Lade das Bild
image = skimage.io.imread('./bilder-Wk08/lake.png')

# Wende den Gauß-Filter an
smoothed_image = skimage.filters.gaussian(image, sigma=2)

#Konvertiere das Bild in ein Grauwertbild


# Berechne die Gradientenstärke mit den Sobel-Operatoren

gradient_magnitude = skimage.filters.sobel(smoothed_image)

#Plotte das Ergebnis
plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')
plt.title('Gradientenstärke')
plt.show()


```

Die stärksten Gradientenveränderung sind weiß dargestellt. Besonders die Übergänge zu Wasser sind stark zu erkennen.
