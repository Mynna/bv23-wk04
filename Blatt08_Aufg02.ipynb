## Aufgabe 2

```python
#Pakete
import numpy as np
import skimage.io
import skimage.filters
import skimage.color
import matplotlib.pyplot as plt

import skimage.exposure
```

## Aufg 2.1

```python


# Lade das Bild
image = skimage.io.imread('./bilder-Wk08/lake.png')

# Wende den Gau-Filter an
# smoothed_image = skimage.filters.gaussian(image, sigma=2)
image = skimage.filters.gaussian(image, sigma=2)

# Entferne den Alphakanal, falls vorhanden, sonst ergibt ein ValueError
if image.shape[2] == 4:
    image = image[:, :, :3]

# Wandle das Bild in den Graustufenbereich um
gray_image = skimage.color.rgb2gray(image)

# Berechne die Gradientenstrke mit den Sobel-Operatoren
gradient_magnitude = skimage.filters.sobel(gray_image)

#Plotte das Ergebnis
plt.imshow(gradient_magnitude, cmap='gray')
plt.axis('off')
plt.title('Gradientenstrke')
plt.show()


```

Die strksten Gradientenvernderung sind wei dargestellt. Besonders die bergnge zu Wasser sind stark zu erkennen.

## Aufg 2.2 

```python

# Lade das Bild
image = skimage.io.imread('./bilder-Wk08/lake-weich.png')

# Entferne den Alphakanal, falls vorhanden
if image.shape[2] == 4:
    image = image[:, :, :3]

# Wandle das Bild in den HSV-Farbraum um
hsv_image = skimage.color.rgb2hsv(image)

# Extrahiere den Hue-Kanal
hue_channel = hsv_image[:, :, 0]

# Wende die nicht-lineare Transformationsfunktion an
transformed_image = np.power(hue_channel, 0.8)

# Strecke den Kontrast des transformierten Bildes
stretched_image = skimage.exposure.rescale_intensity(transformed_image, out_range=(0, 1))

# Skaliere das Bild auf den Bereich von 0 bis 255
scaled_image = skimage.exposure.rescale_intensity(stretched_image, out_range=(0, 255))

# Konvertiere das Bild in den Datentyp uint8
final_image = scaled_image.astype(np.uint8)

# Zeige das Ergebnis
plt.imshow(final_image, cmap='RdYlGn_r')
plt.axis('off')
plt.title('Transformiertes Bild')
plt.show()
```
