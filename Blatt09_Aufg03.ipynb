```python
# Pakete
import numpy as np
import skimage.io
import skimage.filters
import scipy.ndimage
import matplotlib.pyplot as plt

```

```python

# Aufgabe 3.2
laplace_kernel = np.array([[0, -1, 0], [-1, 4, -1], [0, -1, 0]])
gaussian_sigma = 2

combined_filter = skimage.filters.gaussian(laplace_kernel, sigma=gaussian_sigma)

# Aufgabe 3.3
image = skimage.io.imread('./bilder-Wk09/einstein.png') / 255.0  # Lade das Bild und skaliere den Wertebereich auf [0, 1]

filtered_image = scipy.ndimage.convolve(image, combined_filter)

# Plotte das Originalbild und das gefilterte Bild
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(image, cmap='gray')
axes[0].set_title('Originalbild')
axes[0].axis('off')

axes[1].imshow(filtered_image, cmap='gray')
axes[1].set_title('Gau-Laplace-gefiltertes Bild')
axes[1].axis('off')

plt.show()
```
